### Packages

```python
import pandas as pd
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import statsmodels.graphics.tsaplots as sgt 
import statsmodels.tsa.stattools as sts 
from statsmodels.tsa.seasonal import seasonal_decompose
import statsmodels.api as sm
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
import warnings
import joblib
warnings.filterwarnings("ignore")
```

### Loading Data

```python
raw_df = pd.read_csv('./electricity_data.csv')
```

```python
raw_df.head()
```

## Transforming Data

```python
df=raw_df.iloc[:,0:2]
df.rename(columns={'Unnamed: 0':'Date','United States : all sectors':'Power'},inplace=True)
df.Date = pd.to_datetime(df.Date, dayfirst = True)
df.set_index("Date", inplace=True)
df=df.asfreq('MS')
```

```python
df.isna().sum()
```

## Data Exploration

```python
df.describe()
```

```python
df.Power.plot(figsize=(20,8),c='green')
plt.title('Total Monthly Power Genrated in MW (US)',fontsize=(18))
plt.ylabel('Power in MW',fontsize=15)
plt.xlabel('Date',fontsize=15)
```

## Seasonality

```python
s_dec_additive = seasonal_decompose(df.Power, model = "additive")
s_dec_additive.plot()
plt.show()
```

```python
s_dec_multiplicative = seasonal_decompose(df.Power, model = "multiplicative")
fig, (ax1,ax2,ax3) = plt.subplots(3,1, figsize=(15,8))
s_dec_multiplicative.trend.plot(ax=ax1,ylabel = "trend")
s_dec_multiplicative.resid.plot(ax=ax2,ylabel = "seasoanlity")
s_dec_multiplicative.seasonal.plot(ax=ax3,ylabel = "residual")
plt.show()
```

* Trend Shows that power generated increased from 2000 to 2008
* Seasonality is there in the data

### Autocorrelation 

Autocorrelation represents the degree of similarity between a given time series and a lagged version of itself over successive time intervals.

```python
sgt.plot_acf(df.Power, lags = 50, zero = False)
plt.title("ACF Power", size = 24)
plt.show()
```

* Every 11th and 12th is Significatant
* It is due Seasonality in data

### Partial Autocorrelation Function (PACF)

```python
sgt.plot_pacf(df.Power, lags = 50, zero = False, method = ('ols'))
plt.title("PACF Power", size = 24)
plt.show()
```

* Strong Correlation 1,10,11 and 12 lag

### Stationarity Test

```python
sts.adfuller(df.Power)
```

* P value is less than 0.05; So we can say that the data is Non-stationary

### Splitting Model

```python
size = int(len(df)*0.85)
train, test = df.iloc[:size], df.iloc[size:]
```

### Time Series Modelling

```python
model_arma = sm.tsa.arima.ARIMA(train.Power, order=(12,0,7))
results = model_arma.fit()
start,end='2019-01-01', '2022-03-01'

pred = results.predict(start = start, end = end)
pred[start:end].plot(figsize = (20,5), color = "red")
test.Power[start:end].plot(color = "blue")
plt.title("Predictions vs Actual", size = 24)
plt.ylabel('Power in MW',fontsize=15)
plt.xlabel('Date',fontsize=15)
plt.legend(['Prediction','Actual'])
plt.show()
```

```python
print('=============================================')
print('     Time Series Prediction EValuation       ')
print('=============================================')
print('\nR^2 Score',r2_score(test.Power,pred),'\n')
print('Mean Squared Error:',mean_squared_error(test.Power,pred).round(2),'\n')
print('Mean Squared Error:',mean_absolute_error(test.Power,pred).round(2),'\n')
```

### Forecasting Future Dates

```python
model_arma_full = sm.tsa.arima.ARIMA(df.Power, order=(12,0,7))
res = model_arma_full.fit()
forcast = res.predict(start = '2022-04-01', end = '2023-04-01')
forcast.plot(figsize = (20,5), color = "green")
plt.title("Forcasted Power in MW", size = 24)
plt.show()
```

### Saving Model

```python
joblib.dump(res,'model.joblib')
```

### Forecasting Future Power Genartion
